```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
team_data = pd.read_csv('stats.csv')
team_data.head()
```

```python
team_data.describe()
```

```python
team_data[['wins', 'losses', 'goals', 'total_yel_card', 'total_red_card']].median()
```

```python
top_6_teams = team_data[team_data['team'].isin(['Manchester United', 'Chelsea', 'Liverpool', 'Arsenal', 'Tottenham Hotspur', 'Manchester City'])][['team', 'wins', 'season']].reset_index(drop=True)
top_6_teams
```

```python
plt.figure(figsize=(12, 8))

# Loop through each team and plot their data
for team in top_6_teams['team'].unique():
    top_6_team_specific = top_6_teams[top_6_teams['team'] == team]
    plt.plot(top_6_team_specific['season'], top_6_team_specific['wins'], marker='o', label=team)

# Add labels and title
plt.xlabel('Season')
plt.ylabel('Wins')
plt.title('Season Wins for Each Team')
plt.legend(title='Team')
plt.grid(True)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show plot
plt.tight_layout()
plt.show()
```

```python
manchester_united = team_data[team_data.team == 'Manchester United'][['team', 'wins', 'losses','season', 'clean_sheet']].reset_index(drop = True)
manchester_united.head()
```

```python
# we know that in a premier league season there are 38 games 
n = 38.0
manchester_united['win_percentage'] = manchester_united['wins']/n
manchester_united['clean_sheet_percentage'] = manchester_united['clean_sheet']/n
manchester_united
```

```python
# Plotting the win percentage and clean sheet percentage
plt.figure(figsize=(10,6))
plt.plot(manchester_united['season'], manchester_united['win_percentage'], label='Win Percentage', marker='o')
plt.plot(manchester_united['season'], manchester_united['clean_sheet_percentage'], label='Clean Sheet Percentage', marker='o')

# Adding labels and title
plt.xlabel('Season')
plt.ylabel('Percentage')
plt.title('Comparison of Win Percentage and Clean Sheet Percentage')
plt.xticks(rotation=45)
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```
